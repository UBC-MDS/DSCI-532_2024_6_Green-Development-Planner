# Missing Values

```python
# Count the number of missing values in each column
missing_values_count = data.isnull().sum()

# Sort the missing values count in descending order
missing_values_count_sorted = missing_values_count.sort_values(ascending=False)

# Plotting the missing values
plt.figure(figsize=(10, 6))
missing_values_count_sorted.plot(kind='barh')  # Flipping the axes by using 'barh' for horizontal bars
plt.title('Number of Missing Values per Column')
plt.xlabel('Number of Missing Values')  # Flipping x and y labels
plt.ylabel('Columns')
plt.show()
```
