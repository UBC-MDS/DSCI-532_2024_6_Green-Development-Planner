# Missing Values

```python
# Count the number of missing values in each column
missing_values_count = data.isnull().sum()

# Sort the missing values count in descending order
missing_values_count_sorted = missing_values_count.sort_values(ascending=False)

# Plotting the missing values
plt.figure(figsize=(10, 6))
missing_values_count_sorted.plot(kind='barh')  # Flipping the axes by using 'barh' for horizontal bars
plt.title('Number of Missing Values per Column')
plt.xlabel('Number of Missing Values')  # Flipping x and y labels
plt.ylabel('Columns')
plt.show()
```

# Canada Statistics

```python
canada_rows = data[data['Entity'].str.contains('Canada')]
canada_rows
```

```python
numerical_cols = ['Access to electricity (% of population)',
                  'Access to clean fuels for cooking',
                  'Renewable-electricity-generating-capacity-per-capita',
                  'Financial flows to developing countries (US $)',
                  'Renewable energy share in the total final energy consumption (%)',
                  'Electricity from fossil fuels (TWh)', 'Electricity from nuclear (TWh)',
                  'Electricity from renewables (TWh)',
                  'Low-carbon electricity (% electricity)',
                  'Primary energy consumption per capita (kWh/person)',
                  'Energy intensity level of primary energy (MJ/$2017 PPP GDP)',
                  'Value_co2_emissions_kt_by_country',
                  'Renewables (% equivalent primary energy)',
                  'gdp_growth',
                  'gdp_per_capita']
```

```python
# Assuming 'year' is a column in your DataFrame
year_variable = canada_rows['Year']

# List of numerical columns
num_plots = len(numerical_cols)

# Create subplots
fig, axs = plt.subplots(num_plots, 1, figsize=(7, 3*num_plots))

# Plot each numerical column against 'year' on separate subplots
for i, column in enumerate(numerical_cols):
    axs[i].plot(year_variable, canada_rows[column])
    axs[i].set_xlabel('Year')
    axs[i].set_title(f'Line Plot of {column}')

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```
